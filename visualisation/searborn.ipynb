```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = pd.read_csv('data/titanic.csv')
titanic.head()
```

```python
# histogram of the passengers age
plt.figure()
plt.hist(titanic.Age.dropna(), bins=25)
```

```python
# Kernel density estimate
plt.figure()
sns.kdeplot(titanic.Age.dropna(), shade=True)
```

```python
# Violin Plot
plt.figure()
sns.violinplot(titanic.Age.dropna(), titanic.Sex)
```

```python
# cumulative distribution funtion
plt.figure()
sns.kdeplot(titanic.Age.dropna(), cumulative=True)
```

```python
#
# Auto
#

auto = pd.read_csv('data/auto.csv')
auto.describe()
```

```python
sns.pairplot(auto, x_vars=['year','weight'],y_vars='mpg', size=7,aspect=0.7,kind='reg')
```

```python
# Kernel density estimate
sns.kdeplot(auto.year, auto.mpg, shade=True);
```

```python
# joint plot
with sns.axes_style("white"):
    sns.jointplot("weight", "mpg", auto, kind="kde");
```

```python
# Distribution plot
sns.distplot(auto.weight)
```
